```python
# 확인 및 분리
file = '/Users/chuck/Desktop/3-1/Solar-Terrestrial Physics I/우주환경1 과제/GOES-15 p27e csv/total/g15_epead_p27e_32s_2011.txt'
with open(file, 'r') as F:
    data = F.readlines()

time_arr, p3e_flux, p4e_flux, p5e_flux, p6e_flux, p7e_flux = [[], [], [], [], [], []]
p3e_flag, p4e_flag, p5e_flag, p6e_flag, p7e_flag = [[], [], [], [], []]

for line in data:
    line = line.strip()
    column = line.split()
    time_arr.append(column[0] + ' ' +column[1])
    p3e_flag.append(float(column[5]))
    p3e_flux.append(column[7])
    p4e_flag.append(float(column[8]))
    p4e_flux.append(column[10])
    p5e_flag.append(float(column[11]))
    p5e_flux.append(column[13])
    p6e_flag.append(float(column[14]))
    p6e_flux.append(column[16])
    p7e_flag.append(float(column[17]))
    p7e_flux.append(column[19])

print(time_arr[0:5])
#print(p3e_flag)
#print(p3e_flux)
#print(len(time_arr))
#print(len(p3e_flag))
print(len(data))
```

```python
# 이상 데이터가 있는지 확인

error_ind = []

for i in range(len(p3e_flag)):
    if p3e_flag[i] != 0.0:
        error_ind.append(i)

print(error_ind)
print(len(error_ind))
```

```python
# dataframe

import pandas as pd # type: ignore
data = pd.DataFrame({'Time':time_arr, 'P3E_FLUX':p3e_flux, 'P4E_FLUX':p4e_flux, 'P5E_FLUX':p5e_flux, 'P6E_FLUX':p6e_flux, 'P7E_FLUX':p7e_flux})
print(data, type(data))
```

```python
# Time 열을 datetime으로 변경, '숫자' 자료형 부여
data_p = data.copy()
data_p['Time'] = pd.to_datetime(data_p['Time'])
data_p['P3E_FLUX'] = pd.to_numeric(data_p['P3E_FLUX'])
data_p['P4E_FLUX'] = pd.to_numeric(data_p['P4E_FLUX'])
data_p['P5E_FLUX'] = pd.to_numeric(data_p['P5E_FLUX'])
data_p['P6E_FLUX'] = pd.to_numeric(data_p['P6E_FLUX'])
data_p['P7E_FLUX'] = pd.to_numeric(data_p['P7E_FLUX'])
print(data_p)
print(data_p.dtypes)
```

```python
# 데이터 plot
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plt.plot(data_p['Time'], data_p['P3E_FLUX'], label='11.6 MeV')
plt.title('Time Series p3e_flux', fontsize=20)
plt.xlabel('Time', fontsize=20)
plt.ylabel('Proton Flux', fontsize=20)
plt.legend(loc = 'upper right', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
# 보간을 위한 데이터 프레임 생성

data_interp = data_p.copy()
print(data_interp.dtypes)

print(data_interp['P3E_FLUX'][0:10])
```

```python
# 데이터 보간

import numpy as np
import warnings
warnings.filterwarnings('ignore')

for i in error_ind:
    data_interp['P3E_FLUX'][i] = np.nan
    data_interp['P4E_FLUX'][i] = np.nan
    data_interp['P5E_FLUX'][i] = np.nan
    data_interp['P6E_FLUX'][i] = np.nan
    data_interp['P7E_FLUX'][i] = np.nan
data_interp.interpolate(method='linear', inplace=True)

print(data_interp['P3E_FLUX'][0:10])
```

```python
# 결측치 제거 후 확인

for column_name in (list(data_interp.columns)[1:]):
    for i in error_ind:
        print(data_interp[column_name][i])
        print(data_p[column_name][i])

print(data_interp['P3E_FLUX'][0:10])
```

```python
# 보간 후 데이터 plot

plt.figure(figsize=(50, 6))
plt.plot(data_interp['Time'][:], data_interp['P3E_FLUX'][:], label='11.6 MeV', c='b')
#plt.plot(data_interp['Time'][:], data_interp['P4E_FLUX'][:], label='30.6 MeV', c='r')
#plt.plot(data_interp['Time'][:], data_interp['P5E_FLUX'][:], label='63.1 MeV', c='g')
#plt.plot(data_interp['Time'][:], data_interp['P6E_FLUX'][:], label='165 MeV', c='yellow')
#plt.plot(data_interp['Time'][:], data_interp['P7E_FLUX'][:], label='433 MeV', c='black')
plt.title('Time Series - Proton flux in 2011', fontsize=30)
plt.xlabel('Time', fontsize=30)
plt.ylabel('Proton Flux', fontsize=30)
plt.yticks(fontsize = 25)
plt.xticks(fontsize=25)
plt.grid(True)
plt.legend(fontsize=30)
```

```python
# 데이터 resampling

data_res = data_interp.copy()
data_res = data_res.set_index('Time')
data_res = data_res.resample('1T').mean()
```
